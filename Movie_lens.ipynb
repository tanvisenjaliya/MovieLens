# Merging csv files from movie lens dataset

```python
import pandas as pd
```

```python
mnames = ['movie_id','movie_Name','Genres']
movies = pd.read_table(r"C:\Users\ta20111785\Extras\projects\MovieLens\movies.dat", sep='::', header=None,names=mnames,engine='python')

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table(r"C:\Users\ta20111785\Extras\projects\MovieLens\ratings.dat", sep='::', header=None, names=rnames,engine='python')

unames = ['user_id','Gender','Age','preference','5col']
users = pd.read_table(r"C:\Users\ta20111785\Extras\projects\MovieLens\users.dat",sep='::', header=None, names=unames,engine='python')
```

```python
Merged_movies_ratings = pd.merge(movies,ratings,on="movie_id")
Merged_movies_ratings
```

```python
final_merge = pd.merge(users,Merged_movies_ratings, on="user_id")
final_merge
```

```python
df = pd.DataFrame(final_merge)
```

```python
df.to_csv('movies_integrated.csv')
```
